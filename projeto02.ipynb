```python
import pandas as pd
```

```python
tabela = pd.read_csv("Clientes.csv",sep = ";",encoding='latin1')
display(tabela)
```

```python
tabela.drop(columns=['Unnamed: 8'])
```

```python
tabela = tabela.drop(columns=['Unnamed: 8'])
tabela.sort_values(by ='Experiência Trabalho')
```

```python
somaSalarioAnual = tabela['Salário Anual (R$)'].sum()
print(f'{somaSalarioAnual}')
```

```python
print(tabela.info())
```

```python
tabela["Salário Anual (R$)"] = pd.to_numeric(tabela["Salário Anual (R$)"],errors = "coerce")
print(tabela.info())
```

```python
display(tabela[tabela["Profissão"].isna()])
```

```python
tabela = tabela.dropna()
print(tabela.info())
```

```python
somaSalarioAnual = tabela['Salário Anual (R$)'].sum()
print(f'R$ {somaSalarioAnual}')
```

```python
#25,50,75%  organiza a tabela de acordo com a quantidade de itens naquela coluna
#std desvio padrao
display(tabela.describe())
```

```python
import plotly.express as px
grafico  = px.histogram(tabela,x = "Salário Anual (R$)" , y = "Nota (1-100)")
grafico.show()
```
